Task 1: Load and explore dataset

```python
import pandas as pd
df = pd.read_csv('iris_data.csv')
print(df)
```

Display the first few rows of the dataset using .head() to inspect the data.

```python
df.head(5)
```

Explore the structure of the dataset by checking the data types and any missing values.

```python
df.isnull().sum()
```

Clean the dataset by either filling or dropping any missing values.

```python
# Since there are no missing values, this step will not modify the dataset
df = df.dropna()  # Drop rows with missing values (if any)
```

Task 2: Basic Data Analysis
Compute the basic statistics of the numerical columns (e.g., mean, median, standard deviation) using .describe().

```python
df.describe()
```

Perform groupings on a categorical column (for example, species, region, or department) and compute the mean of a numerical column for each group.

```python
df.groupby('Iris-setosa').mean()
```

identify any patterns or interesting findings from your analysis.

```python
# Count the number of samples for each species
species_counts = df['Iris-setosa'].value_counts()
print("Species distribution:\n", species_counts)

# Compute the mean of numerical columns for each species
species_means = df.groupby('Iris-setosa').mean()
print("\nMean values for each species:\n", species_means)
```

```python
# Rename the columns to meaningful names
df.columns = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', 'Species']
print(df.head())
```

Task 3: Data Visualization
Line chart,
Bar chart,
Histogram, and 
Scatter plot 

Scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

Line chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

Bar chart

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Species', y='SepalLengthCm', hue='Species')
plt.title('Average Sepal Length by Species')    
plt.xlabel('Species')
plt.ylabel('Average Sepal Length (cm)')
plt.legend(title='Species')
plt.show()
```

Histogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='SepalLengthCm', hue='Species', multiple='stack')
plt.title('Distribution of Sepal Length by Species')    
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')    
plt.legend(title='Species')
plt.show()
```
